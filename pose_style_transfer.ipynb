```python
from api import Pose2Pose
from PIL import Image

p2p = Pose2Pose(pretrained=True)

```

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
# Load images using OpenCV
condition_cv = cv2.imread('/content/pose-transfer/api-test/condition.jpg')
reference_cv = cv2.imread('./api-test/target_pose_reference.jpg')

condition_rgb = cv2.cvtColor(condition_cv, cv2.COLOR_BGR2RGB)
reference_rgb = cv2.cvtColor(reference_cv, cv2.COLOR_BGR2RGB)


condition_pil = Image.fromarray(condition_rgb)
reference_pil = Image.fromarray(reference_rgb)
generated = p2p.transfer_as(condition_pil, reference_pil)
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(condition_pil)
plt.title('Condition Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(reference_pil)
plt.title('Reference Pose Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(generated)
plt.title('Generated Output Image')
plt.axis('off')

plt.show()
```
